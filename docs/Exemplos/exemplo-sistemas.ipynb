# Points

```julia
A_p1 = [Point2D(1,2), Point2D(3,4)]
Draw(A_p1)
```

```julia
C_p1 = Point2D(11,16)
Draw(C_p1, false)
```

```julia
B_p1 = A_p1\C_p1
```

```julia
A_p2 = [ Point3D(-1.0, -11.0, -3.0), Point3D(1.0, 1.0, 0.0 ), Point3D(2.0, 5.0, 1.0 ) ]
Draw(A_p2)
```

```julia
C_p2 = Point3D(7.0, 6.0, 0.0)
Draw(C_p2, false)
```

```julia
B_p2 = A_p2\C_p2
```

# Setas

```julia
A_s1 = [Arrow2D(5,2), Arrow2D(-2,-3)]
Draw(A_s1)
```

```julia
C_s1 = Arrow2D(11,-11)
Draw(C_s1, false)
```

```julia
B_s1 = A_s1\C_s1
```

```julia
A_s2 = [ Arrow3D(-1.0, -11.0, -3.0), Arrow3D(1.0, 1.0, 0.0 ), Arrow3D(2.0, 5.0, 1.0 )  ]
Draw(A_s2)
```

```julia
C_s2 = Arrow3D(7, 6, 0)
Draw(C_s2, false)
```

```julia
B_s2 = A_s2\C_s2
```

# Signal

```julia
A_sg1 = [SignalBuildU(x -> cos(100*x), 0:0.1:10), SignalBuildU(x -> 10*sin(x), 0:0.1:10)]
Draw(A_sg1)
```

```julia
b_temp = [2, 3]
C_sg1 = A_sg1 * b_temp
Draw(C_sg1, false)
```

```julia
B_sg1 = A_sg1\C_sg1
```
