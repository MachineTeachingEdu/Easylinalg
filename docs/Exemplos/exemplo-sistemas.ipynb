# Points

```julia
p1 = [Point2D(1,2), Point2D(3,4)]
Draw(p1)
```

```julia
b = [2, 3]
c = p1 * b
Draw(c)
```

```julia
p1\c
```

```julia
p2 = [ Point3D(-1.0, -11.0, -3.0), Point3D(1.0, 1.0, 0.0 ), Point3D(2.0, 5.0, 1.0 )  ]
Draw(p2)
```

```julia
b = [1, 2, 3]
c = p2 * b
Draw(c)
```

```julia
p2\c
```
