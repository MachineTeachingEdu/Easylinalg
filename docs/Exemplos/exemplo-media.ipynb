# Points

```julia
p1 = Point2D(2, 2)
p2 = Point2D(5, 9)
p3 = Point2D(5, 1)
p4 = Point2D(10, 2)
p5 = Point2D(-5, 10)
p6 = Point2D(7, 9)
Draw([p1,p2,p3,p4,p5,p6])
```

```julia
p_media = (p1+p2+p3+p4+p5+p6) / 6
Draw(p_media, false)
```

```julia
p_3d_1 = Point3D(1, 1, 1)
p_3d_2 = Point3D(7, 10, -5)
p_3d_3 = Point3D(-4, 2, 11)
p_3d_4 = Point3D(10, 5, 3)
p_3d_5 = Point3D(-2, -3, -8)
p_3d_6 = Point3D(4, 6, -9)
p_3d_7 = Point3D(2, -2, 6)
p_3d_8 = Point3D(7, 4, 7)
Draw([p_3d_1,p_3d_2,p_3d_3,p_3d_4,p_3d_5,p_3d_6,p_3d_7,p_3d_8])
```

```julia
p_3d_medio = (p_3d_1+p_3d_2+p_3d_3+p_3d_4+p_3d_5+p_3d_6+p_3d_7+p_3d_8)/8
Draw(p_3d_medio, false)
```

# Signal

```julia
sg1 = SignalBuildU(x -> cos(100*x), -10:0.1:10)
sg2 = SignalBuildU(x -> 10*sin(x), -10:0.1:10)
Draw([sg1, sg2])
```
