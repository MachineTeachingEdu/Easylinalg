## Transormação Linear



```julia
p1 = Point2D(1,2)
p2 = Point2D(3,2)
p3 = Point2D(2,4)
Draw([p1,p2,p3])
```

```julia
B = [1 0;
     0 -1]
p1 = p1 * B
p2 = p2 * B
p3 = p3 * B
Draw([p1, p2, p3])
```

```julia
A_3d = Point3D(1,2,3)
Draw(A_3d)
```

```julia
B_3d = [1 0 0;
     0 -1 0;
     0 0 1]
C_3d = A_3d * B_3d
Draw(C_3d, false)
```

# Setas

```julia
A_arrow = Arrow2D(1,2)
Draw(A_arrow)
```

```julia
B_arrow = [1 0;
     0 -1]
C_arrow = A_arrow * B_arrow
Draw(C_arrow, false)
```

```julia
A_arrow_3d = Arrow3D(1,2,3)
Draw(A_arrow_3d)
```

```julia
B_arrow_3d = [1 0 0;
              0 -1 0;
              0 0 1]
C_arrow_3d = A_arrow_3d * B_arrow_3d
Draw(C_arrow_3d, false)
```
